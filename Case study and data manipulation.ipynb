2\. Write a Python function called `getDfSummary()` that does the following:
- Takes as input a data frame
- For each variable in the data frame calculates the following features:
  - `number_nan` to count the number of missing not-a-number values
  - Ignoring missing, NA, and Null values:
    - `number_distinct` to count the number of distinct values a variable can take on
    - `mean`, `max`, `min`, `std` (standard deviation), and `25%`, `50%`, `75%` to correspond to the appropriate percentiles
- All of these new features should be loaded in a new data frame. Each row of the data frame should be a variable from the input data frame, and the columns should be the new summary features.
- Returns this new data frame containing all of the summary information

Hint: The pandas `describe()` [(manual page)](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.describe.html) method returns a useful series of values that can be used here.

```python

def getDfSummary(input_data):
    import pandas as pd
    import numpy as np
    from pandas import DataFrame
    
    # Place your code here
    #Create column name
    column_name = ['number_nan','number_distinct','mean','max','min','std','25%','50%','75%']
    
    #Create row name
    row_name = input_data.columns
    
    #Create empty data frame
    empty = np.empty((len(row_name),len(column_name)),dtype=np.float)
    empty[:] = np.nan
    
    output_data = pd.DataFrame(empty)
    
    #Insert column names
    output_data.columns = column_name
    
    #Insert row names
    index = pd.Index(row_name)
    output_data = pd.DataFrame(output_data, index=index)
    
    #Insert into the first column: number_nan
    describe = input_data.describe()
    for i in range(0, input_data.shape[1]):
         output_data.iloc[i,0]= input_data.shape[0] - describe.iloc[0,:][i]
      
    #Insert into the second column:number_distinct
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,1]=len(ads.iloc[:,i].unique())
        
    for i in range(0, input_data.shape[1]):
        if output_data.iloc[i,0] != 0:
            output_data.iloc[i,1] -= 1
    
    #Insert into the thrid column: mean
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,2]= describe.iloc[1,:][i]
        
    #Insert into the fourth column: max
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,3]= describe.iloc[7,:][i]
        
    #Insert into the fifth column: min
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,4]= describe.iloc[3,:][i]
    
    #Insert into the sixth column: std
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,5]= describe.iloc[2,:][i]
    
    #Insert into the seventh column: 25%
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,6]= describe.iloc[4,:][i]
        
    #Insert into the eigth column: 50%
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,7]= describe.iloc[5,:][i]
    
    #Insert into the eigth column: 75%
    for i in range(0, input_data.shape[1]):
        output_data.iloc[i,8]= describe.iloc[6,:][i]
        
        
        
        
    return output_data               
                     
```
